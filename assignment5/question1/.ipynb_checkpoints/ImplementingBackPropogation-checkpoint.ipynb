```python
# Importing the required libraries
import os
import matplotlib.pyplot as plt
import PIL
import numpy as np
import math
from sklearn.decomposition import PCA
```

```python
'''
Helper class for holding dataset
    Attributes
        * data_type : holds the type of the dataset (train/test)
        * N : number of image files in the dataset
        * dataset : a numpy matrix of dimensions = (Number of images) x (Dimensionality of each image) 
        * labels : List storing the class labels of each of the image 0 for sad images 1 for happy images.
'''

class Dataset:
    
    def __init__(self,dataset_type="Data/emotion_classification/train/"):
        
        self.data_type = dataset_type
        self.files = os.listdir(dataset_type)
        self.N = len(self.files)
        self.dataset = None
        self.labels = []
    

    # Constructs the data matrix X of dimensions (Number of images x Dimension of 1 image vector)
    def getDataset(self):
        
        # Iterate throigh
        for file in self.files:
            # Get the Image corresponding to the data point
            pilimage = PIL.Image.open(self.data_type + file)
            img_class = file.split(".")[1]
            
            # If the class of the image is sad append a label 0
            if(img_class=='sad'):
                self.labels.append(0)
            else:
                self.labels.append(1)
    
            pilimage.seek(0)
            # Convert the image to array
            tmp = pilimage.convert()

            # Convert the image to a numpy array
            current_image = np.asarray(tmp)
            # Normalize the pixel values
            current_image = current_image/255.0
            # Vectorize the image
            current_image = current_image.reshape((1,-1))
            
            # Append along the dimension 0 for the data points which are coming 
            # to create the feature matrix.
            if self.dataset is None:
                self.dataset = current_image
            else:
                self.dataset = np.append(self.dataset,current_image,axis=0)
    
        return self.dataset,self.labels
```

```python
# Get the training and test data set along with the true labels
train , test = Dataset("Data/emotion_classification/train/") ,Dataset("Data/emotion_classification/test/")
train_set,train_labels = train.getDataset()
test_set,test_labels = test.getDataset()
```

```python
# Perform PCA using SkLearns PCA Implementaion
# The principal components are evaluated with respect to the training set
# both the training and test set are projected on to these principal components.

pca = PCA(n_components=12)
train_set = pca.fit_transform(train_set)
test_set = pca.transform(test_set)
```

### Neural Network

- Feed Forward Phase
    -  Input = x     
    - Dimensions of x =  12x1
    - $in1 = W^1x+b^1$ 
    - $out1 = ReLU(in1)$
    - $in2 = W^2out1 + b^2$ 
    - $out2 = Softmax(in2)$

    $out2$ would be class probabilities which is a 2x1 vector where $out2[0][0]$ represent the probability of sad class while $out2[1][0]$ represent the probability of happy class.

- Loss Calculation
    - Cross Entropy loss used $- \sum_{n=1}^{N} \sum_{k}^{numClasses} (t_{n}[k]*log(out2_{n}[k]))$
    
    where t_{n}[k] is the probability of the nth data point belonging to class k

- Gradients for a particular data point 

    -  $\frac{\partial E}{\partial in2[k]} = \delta^{2}[k] = (out2[k]-t[k])$ [Reference: https://www.youtube.com/watch?v=5-rVLSc2XdE ] 
    
    - $\frac{\partial E}{\partial W^{(2)}_{jk}} = \delta^{2}[k] * out^{1}[j]$

    - $\frac{\partial E}{\partial b^{k}} = \delta^{2}[k]$

    - $\frac{\partial E}{\partial in1[j]} = \delta^{1}[j] = \sum^{}_{k} W^{(2)}_{kj} \delta^{(2)}_{k}$  if $in1[j]$ is positive 
        - if $in1[j]$ is negative $\frac{\partial E}{\partial in1[j]} = 0$

        - Derivative of ReLU is 1 for positive while 0 for negative inputs Hence the formula came

    - $\frac{\partial E}{\partial W^{(1)}_{ij}} = \delta^{1}[j] * input[i] $
    
    - $\frac{\partial E}{\partial b^{j}} = \delta^{1}[j]$
    

```python
# Define the class for the neural network

class NN:
    """
        Attributes:
        ----------------------

            input_dim: The dimension  of the input vector 
            hidden_dim : Number of neurons in the hidden layer
            num_classes: Number of classes for classification
            learning_rate : The rate at which descent happens

            Neural Network Parameters
            ----------------------------
                self.weight1 = W^1 in the neural network
                self.weight2 = W^2 in the neural network
                self.bias1 = b^1 in the neural network
                self.bias2 = b^2 in the neural network
        Methods:
        ---------------------
            softmax: computes the softmax activation function
            relu : computes the relu activation function
            forward: feed forward phase for a neural network
            forward_with_info : feed forward phase with info of intermediate outputs
            one_hot_encode: convert the class label into a 1 hot encoded vector which is used for error calculation.
            test : Get the accuracy for the test set and labels given as input
            plot_losses: Function to plot the loss as a function of the number of epochs
            train: Train the neural network using the input dataset and labels 

    """
    def __init__(self,input_dim,hidden_dim,num_classes=2,learning_rate=0.01):
        
        self.input_dim = input_dim
        self.hidden_dim = hidden_dim
        self.num_classes = num_classes
        self.learning_rate = 0.01

        # The parameters of the neural network
        self.weight1 = np.random.randn(hidden_dim,input_dim)
        self.weight2 = np.random.randn(num_classes,hidden_dim)
        self.bias1 = np.random.randn(hidden_dim,1)
        self.bias2 = np.random.randn(num_classes,1)

        # Stores the gradients during each epoch with which 
        # the weights in the neural network would be updated
        self.gradient_weight1 = np.zeros(self.weight1.shape)
        self.gradient_weight2 = np.zeros(self.weight2.shape)
        self.gradient_bias1 = np.zeros(self.bias1.shape)
        self.gradient_bias2 = np.zeros(self.bias2.shape)

    def relu(self,x):
        # ReLU activation function
        return np.maximum(x,0)
    
    def softmax(self,x):
        # Softmax Operation over a vector of shape (No of classes,1)
        return np.exp(x) / np.sum(np.exp(x),axis=0)    

    def forward(self,x):
        
        # Create a copy of the original input
        inp = np.copy(x)

        # Compute W^1x + b^1
        in1 = self.weight1.dot(x) + self.bias1
        # Compute ReLU(W^1x + b^1)
        out1 = self.relu(in1)

        # Compute W^2 out1 + b^2
        in2 = self.weight2.dot(out1)+self.bias2

        # Compute Softmax(W^2 out1 + b^2)
        out2 = self.softmax(in2) 
        
        # Return the output
        return out2
        
    def forward_with_info(self,x):
        # Create a copy of the original input
        inp = np.copy(x)

        # Compute W^1x + b^1
        in1 = self.weight1.dot(x) + self.bias1
        # Compute ReLU(W^1x + b^1)
        out1 = self.relu(in1)

        # Compute W^2 out1 + b^2
        in2 = self.weight2.dot(out1)+self.bias2

        # Compute Softmax(W^2 out1 + b^2)
        out2 = self.softmax(in2)  
        
        # Return all the outputs
        return (in1,out1,in2,out2)
    

    def one_hot_encode(self,label):
        
        # Create an array of zeros of the size of the number of classes
        one_hot_vector = np.zeros(self.num_classes)
        
        # Convert the above array into a vector form
        one_hot_vector = np.expand_dims(one_hot_vector,axis=1)  
        
        # Store 1 in the label given 
        one_hot_vector[label][0] = 1
        
        return one_hot_vector
      
   
    def test(self,x,labels):
        correct = 0         # The number of correctly classified points
        
        for i in range(x.shape[0]):
            inp = np.copy(x[i])
            # Convert the input array to a vector format
            inp = np.expand_dims(inp,axis=1)

            # Get the scores from the neural network.
            output = self.forward(inp)
            # Find the class label of the current input
            label = np.argmax(output,axis=0)
            # Increment the count if the class labels match
            if(label==labels[i]):
                correct+=1
        
        # Accuracy of the classifier
        accuracy = (correct/x.shape[0])*100
        print("Accuracy of the trained Neural Network = " + str(accuracy))
          

    def plot_losses(self,losses):
        plt.xlabel("Epochs")
        plt.title("Epochs vs Losses")
        plt.ylabel("Loss")

        # The values in the x axis
        x = [i for i in range(1,len(losses)+1)]
        plt.plot(x,losses,color='orange')
        plt.show()


    def train(self,x,labels,epochs):
        
        losses = [] # Store the losses averaged out through each epoch
        for e in range(epochs):

            # Reset the gradients to zero before moving on to the input data
            self.gradient_weight1 = np.zeros(self.weight1.shape)
            self.gradient_weight2 = np.zeros(self.weight2.shape)
            self.gradient_bias1 = np.zeros(self.bias1.shape)
            self.gradient_bias2 = np.zeros(self.bias2.shape)

            # Cross Entropy Loss 
            total_loss = 0

            
            for i in range(x.shape[0]):
                inp = np.copy(x[i])
                # Convert the input into vector format
                inp = np.expand_dims(inp,axis=1)
                # Get the label corresponding to the data point
                label = labels[i]
                # Get the one hot vector corresponding to the label.
                true_value = self.one_hot_encode(label)

                # Get all the intermediate outputs from the forward pass
                in1,out1,in2,out2 = self.forward_with_info(inp)

                # Computing the Cross Entropy Loss
                loss_value = true_value*np.log(out2)
                loss = -np.sum(loss_value,axis=0)
                total_loss += loss

                # Compute d(error) / d(W^2 x+b^2) 
                self.del_k = out2 - true_value # Reference 

                
                interjk = np.repeat(out1.transpose(),self.del_k.shape[0],axis=0)
                # Compute d(error) / d(W^2_{jk})
                self.gradient_weight2 += (interjk*self.del_k)
                # Compute d(error) / d(b^2_{k})
                self.gradient_bias2 +=self.del_k

                # Compute delta_j as given in the equation in the markdown above
                self.del_j = np.matmul(self.weight2.transpose(),self.del_k)
                # Applying the derivative of the ReLU activation function which is 1 for positive points and 0 
                # for negative points and multiplying with the intermediate result computed from above.
                self.del_j = np.where(self.del_j>=0,self.del_j,0)


                interij = np.repeat(inp.transpose(),self.del_j.shape[0],axis=0)
                # Compute d(error) / d(W^1_{ij})
                self.gradient_weight1 += (interij*self.del_j)
                # Compute d(error) / d(b^1_{j})
                self.gradient_bias1 += self.del_j
            
            # Get the loss averaged out through the epoch.
            averaged_loss = total_loss / x.shape[0]
            # Append to the losses list which would be used while plotting
            losses.append(averaged_loss[0])

            # Average the gradients computed.
            self.gradient_bias1 = self.gradient_bias1 * (1.0/x.shape[0])
            self.gradient_bias2 = self.gradient_bias2 * (1.0/x.shape[0])
            self.gradient_weight1 = self.gradient_weight1 * (1.0/x.shape[0])
            self.gradient_weight2 = self.gradient_weight2 * (1.0/x.shape[0])
            print("Epoch" + str(e) + " Average loss = " + str(averaged_loss[0]))

            # Perform gradient descent step 
            self.weight1 = self.weight1 - self.learning_rate*self.gradient_weight1
            self.weight2 = self.weight2 - self.learning_rate*self.gradient_weight2
            self.bias1 = self.bias1 - self.learning_rate*self.bias1
            self.bias2 = self.bias2 - self.learning_rate*self.bias2
        
        # Plot losses as a function of the number of epochs.
        self.plot_losses(losses)

        
```

```python
# Neural Network with hidden layer size 10
np.random.seed(134) # To get deterministic results as the weights in the neural network are initialized randomly.

# Create the neural network train it and test for its accuracy
nn10 = NN(input_dim=12,hidden_dim=10)
nn10.train(x = train_set,labels=train_labels,epochs = 30)

print("------------------------------------")
nn10.test(x = test_set,labels=test_labels)
```

```python
# Neural Network with hidden layer size 15
np.random.seed(134) # To get deterministic results as the weights in the neural network are initialized randomly.
nn15 = NN(input_dim=12,hidden_dim=15) 
# Create the neural network train it and test for its accuracy
nn15.train(x = train_set,labels=train_labels,epochs = 30)

print("------------------------------------")
nn15.test(x = test_set,labels=test_labels)
```

**We notice a slight increase in the accuracy when the number of hidden layer neurons is increased from 10 to 15**
