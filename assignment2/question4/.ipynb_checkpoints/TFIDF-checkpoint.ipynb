```python
import numpy as np
import string
import numpy as np
from scipy.stats import multivariate_normal
import math
import matplotlib.pyplot as plt


f = open("movieReviews1000.txt")
sentences = []
labels = []

punctuations =[]
for p in string.punctuation:
    punctuations.append(p)

for line in f.readlines():
    words = line.strip().split(" ")
    labels.append(words[-1].strip())
    
    lowercase_words = []
    for word in words[:-2]:
        lowercase_words.append(word.lower())
    sentences.append(lowercase_words)
```

```python
def preprocess_words(sentences):
    new_sentences = []
    for sentence in sentences:
        words = []
        for word in sentence:
            if word =='':
                continue
            elif word[-1] in '+?:;*&':
                word = word[:-1]
            if '?' in word:
                word = word.replace('?','')
            words.append(word)
        new_sentences.append(words)
    return new_sentences

sentences = preprocess_words(sentences)            
```

```python

#Preprocessing the text data
word_set = []
for sentence in sentences:
    for word in sentence:
        if word not in word_set:
            word_set.append(word)
        
#Set of vocab 
word_set = set(word_set)
#Total documents in our corpus
total_documents = len(sentences)
 
#Creating an index for each word in our vocab.
index_dict = {} #Dictionary to store index for each word
i = 0
for word in word_set:
    index_dict[word] = i
    i += 1

```

```python
#Create a count dictionary
def count_dict(sentences):
    word_count = {}
    for word in word_set:
        word_count[word] = 0
        for sent in sentences:
            if word in sent:
                word_count[word] += 1
    return word_count
 
word_count = count_dict(sentences)
word_count
```

```python
#Term Frequency
def termfreq(document, word):
    N = len(document)
    occurance = len([token for token in document if token == word])
    return occurance/N
```

```python
#Inverse Document Frequency
 
def inverse_doc_freq(word):
    try:
        word_occurance = word_count[word] + 1
    except:
        word_occurance = 1
    return np.log(total_documents/word_occurance)
```

```python
def tf_idf(sentence):
    tf_idf_vec = np.zeros((len(word_set),))
    for word in sentence:
        tf = termfreq(sentence,word)
        idf = inverse_doc_freq(word)
         
        value = tf*idf
        tf_idf_vec[index_dict[word]] = value 
    return tf_idf_vec
```

```python
#TF-IDF Encoded text corpus
vectors = []
for sent in sentences:
    vec = tf_idf(sent)
    vectors.append(vec)
```

```python
tf_idf_matrix = np.array(vectors)
```

```python
tf_idf_matrix.shape
```

```python
# Create the vocabulary out of all the words given
count = 0
vocab = {}
for sentence in sentences:
    for word in sentence:
        word = word.lower()
        if word not in vocab:
            vocab[word] = count
            count+=1
```

```python
doc_frequency = {}
for word in vocab:
    doc_frequency[word] = 0
    for sentence in sentences:
        if word in sentence:
            doc_frequency[word]+=1

idf_vector = np.zeros((len(vocab),1))
for word in doc_frequency:
    idf_vector[vocab[word]] = np.log10((float(1+1000.0))/(1+doc_frequency[word]))
print(idf_vector)
```

```python
tf_idf_matrix = None
vocab_length = len(vocab)

for sentence in sentences:
    
    sentence_length = len(sentence)
    embedding = np.zeros(vocab_length)
    
    for word in sentence:
        embedding[vocab[word]] +=1
    
    embedding = embedding*(1.0/len(sentence))
    
    tf_idf = np.zeros((len(vocab),1))
    for i in range(len(vocab)):
        tf_idf[i] = embedding[i]*idf_vector[i][0]
    if tf_idf_matrix is None:
        tf_idf_matrix = tf_idf.reshape((1,-1))
    else:
        tf_idf_matrix = np.append(tf_idf_matrix,tf_idf.reshape((1,-1)),axis=0)
    #print(embedding[voca])
    #print(embedding)
print(tf_idf_matrix.shape)
for i in range(11):
    print(tf_idf_matrix[0][i])
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=10)
print(tf_idf_matrix.shape)
transformed_data = pca.fit_transform(tf_idf_matrix)
print(transformed_data.shape)
```
