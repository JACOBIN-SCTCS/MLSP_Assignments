```python
import string
import numpy as np
from scipy.stats import multivariate_normal
import math
import matplotlib.pyplot as plt

f = open("movieReviews1000.txt")
sentences = []
labels = []

# Store all the punctuations
punctuations =[]
for p in string.punctuation:
    punctuations.append(p)

# Read all the sentences with their labels stored in labels
for line in f.readlines():
    words = line.strip().split(" ")
    labels.append(words[-1].strip())
    words = words[:-2]
    lower_case_words = []
    for word in words:
        lower_case_words.append(word.lower())
    sentences.append(lower_case_words)
```

```python
sentences[0]
```

```python
def preprocess_text(sentences):
    new_sentences = []
    for sentence in sentences:
        new_words = []
        for word in sentence:
            if word=='':
                continue
            if word[-1] in ['?',';',':','&']:
                word = word[:-1]
            if word=='':
                continue
            new_words.append(word)
        new_sentences.append(new_words)
    return new_sentences

sentences = preprocess_text(sentences)
```

```python
# Create the vocabulary out of all the words given
vocab = {}
cnt = 0
for sentence in sentences:
    for word in sentence:
        if word not in vocab:
            vocab[word] = cnt
            cnt+=1
vocab
```

```python
idf_vector = np.zeros((len(vocab)))
for word in vocab:
    for sentence in sentences:
        if word in sentence:
            idf_vector[vocab[word]] +=1

for i in range(len(idf_vector)):
    idf_vector[i] = np.log(((1000.0)/(idf_vector[i])))
idf_vector
```

```python
tf_idf_matrix = None
for s in sentences:
    tf_term = np.zeros((1,len(vocab)))
    for word in s:
        tf_term[0][vocab[word]]+=1
    
    tf_term = tf_term*(1.0/len(s))
    print(tf_term)
    
    for i in range(len(vocab)):
        tf_term[0][i] *= idf_vector[i]
    
    if tf_idf_matrix is None:
        tf_idf_matrix = tf_term
    else:
        tf_idf_matrix = np.append(tf_idf_matrix,tf_term,axis=0)

print(tf_idf_matrix.shape)
        
```

```python
tokenized_list_of_sentences = sentences
from sklearn.feature_extraction.text import TfidfVectorizer
def identity_tokenizer(text):
    return text

tfidf = TfidfVectorizer(tokenizer=identity_tokenizer,vocabulary=vocab ,lowercase=False,use_idf=True,smooth_idf=False)    
X = tfidf.fit_transform(tokenized_list_of_sentences)

```

```python
class GMM:
    
    def __init__(self,k,data_matrix,max_iter=100):
        self.k = k
        self.d = data_matrix.shape[1]
        self.max_iter = max_iter
        self.data_matrix = data_matrix
        self.colors = ['r','g','b','o','y']
        self.centers = None
        self.cov = None
        self.pi = None
        self.normal_curves = None
    
    '''def normal_equation(self,data, mean, cov):
        d = len(mean)              
        mean = mean.reshape(d, 1)     
        data = data.reshape(d, 1)     

        constant = (2 * np.pi)**d        
        determinant = np.linalg.det(cov)      
        constant = 1.0 / np.sqrt(constant *determinant) 

        vector_t = np.transpose(data - mean)          
        cov_inverse = np.linalg.inv(cov)   
        vector = ( data  - mean)       
        exp_term = -0.5 * (np.dot(vector_t,cov_inverse).dot(vector))

        result = constant * np.exp(exp_term)  
        return result[0][0]'''


    def normal_equation(self,data,mean,covariance):
        
        ans = 1.0/((np.power(2*math.pi,1.0/self.d))*(np.sqrt(np.linalg.det(covariance))))
        data_vector = data.reshape((-1,1)) - mean.reshape((-1,1))
        sigma_inverse = np.linalg.inv(covariance)
        temp_product = np.matmul(data_vector.transpose(),sigma_inverse)
        temp_product = np.matmul(temp_product,data_vector)
        exponential_term = np.exp(temp_product[0][0]*(-1.0/2))
        print(exponential_term)

        probability = exponential_term*ans
        return probability
        
    def initialize(self):
        
        split_points = np.split(self.data_matrix,self.k)
        self.centers = {}
        self.cov = {}
        self.pi = {}
        for i in range(len(split_points)):
            self.centers[i] = np.mean(split_points[i],axis=0)
            #self.cov[i] = np.cov(split_points[i].transpose())
            covariance_matrix = np.cov(split_points[i].transpose()).diagonal()
            diag_cov = np.eye(self.d)
            for k in range(self.d):
                diag_cov[k][k] = covariance_matrix[k]
            self.cov[i] = diag_cov
            self.pi[i] = 1.0/self.k
 
    def plot(self,data_points):
        for point in data_points:
            plt.scatter(point[0],point[1],c=self.colors[point[2]])
        plt.show()
        
    
    def e_step(self):      
        
        points_to_plot = []
        for i in range(len(self.data_matrix)):
            probability = []
            for k in range(self.k):
                prob = self.normal_equation(self.data_matrix[i],self.centers[k],self.cov[k])
                #prob = prob_fn(self.data_matrix[i])
                prob = self.pi[k]*prob
            
                probability.append(prob)
            print(probability)
            max_value = max(probability)
            max_index = probability.index(max_value)
            
            points_to_plot.append((self.data_matrix[i][0],self.data_matrix[i][1],max_index))
        cnt = 0
        for point in points_to_plot:
            if point[2]==1:
                cnt+=1
        print(cnt)
        self.plot(points_to_plot)
    
    def m_step(self):
        pass
    
    def k_means(self):
        pass
    
gmm = GMM(2,transformed_data)
gmm.initialize()
gmm.e_step()
```
