```python
# Importing the required libraries
import os
import matplotlib.pyplot as plt
import PIL
import numpy as np
import math
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.svm import SVC
from mpl_toolkits import mplot3d

```

```python

class Dataset:
    '''
    Helper class for holding dataset
        *Attributes
            * data_type : holds the type of the dataset (train/test) 
            * files : List containing the name of fies in the dataset
            * N : number of image files in the dataset 
            * n_features : Store the dimensions of each image vector
            * emotion : list storing the class associated with each image 0 for happy 1 for sad
            * dataset : a numpy matrix of dimensions = (Number of images) x (Dimensionality of each image) 
    '''
    def __init__(self,dataset_type="train/"):
        
        self.data_type = dataset_type
        self.files = os.listdir(dataset_type)
        self.N = len(self.files)
        self.n_features = []
        self.emotion = []
        self.dataset = None
    
    # Constructs the data matrix X of dimensions (Number of images x Dimension of 1 image vector)
    def getDataset(self):
        
        index = 0
        # Iterate over all the files
        for file in self.files:
            pilimage = PIL.Image.open(self.data_type + file)
            
            img_class = file.split(".")[1]
            if(img_class=='sad'):
                self.emotion.append(1)
            else:
                self.emotion.append(0)
            
            # Convert the PIL Image to a matrix
            pilimage.seek(0)
            tmp = pilimage.convert()    
            current_image = np.asarray(tmp)

            # Flatten the matrix into a vector and append it to the data matrix so 
            # that dimension of dataset is  (number of images * vector size)
            current_image = current_image.reshape((1,-1))
            if self.dataset is None:
                self.dataset = current_image
            else:
                self.dataset = np.append(self.dataset,current_image,axis=0)
    
            self.n_features.append(current_image.shape[1])
            index+=1
        return self.dataset
```

```python
# Obtain the dataset and generate the feature matrix
train = Dataset("Data/emotion_classification/train/")
train_data = train.getDataset()

# Obtain the test dataset and generate the feature matrix
test = Dataset("Data/emotion_classification/test/")
test_data = test.getDataset()

print(train_data.shape)
print(test_data.shape)
```

```python
class PCAHelper:
    '''
        Helper class for performing Principal Component Analysis
        Attributes
        ----------
            dataset : (Number of Datapoints * Dimension of each data point)
            n_components : Number of Principal Components
    '''

    def __init__(self,dataset,n_components):
        self.dataset = dataset
        self.n_components = n_components
        self.projected_dataset = None
        self.pca = PCA(n_components= self.n_components)
    
    # Perfrom PCA on self.dataset
    def perform_pca(self):
        # Get the principal components corresponding to the training dataset
        # and project the data onto these principal components
        self.projected_dataset = self.pca.fit_transform(self.dataset)
        return self.projected_dataset

    # Perform projection of the given dataset into the principal components
    def transform(self,test_dataset):
        # Obtain the test feature matrix after projection onto the principal components
        return self.pca.transform(test_dataset)
```

```python
class LDAHelper:
    
    '''
        Helper class for performing Linear Discriminant Analysis
        Attributes
        ----------
            dataset : (Number of Datapoints * Dimension of each data point)
            labels : The class label for each data point (Number of Data points,)
    '''

    def __init__(self,dataset,labels):
        self.dataset = dataset
        self.projected_dataset = None
        self.lda = LinearDiscriminantAnalysis() # Sklearn's LDA function
        self.labels = np.array(labels)
    
    # Perform LDA on self.dataset with label information in self.labels
    def perform_lda(self):
        # Perform Linear Discriminant Analysis on the training data and get 
        # projection vectors that maximizes the Fischer Ratio
        self.projected_dataset = self.lda.fit_transform(self.dataset,self.labels)
        return self.projected_dataset
    
    
    # Perform classification using the learned LDA model 
    def perform_classification(self,dataset):
        return self.lda.predict(dataset)
    
    # Calculate the accuracy on the test dataset wth label information given by the parameter labels
    # dataset: (No of Data points * Dimension of each point)
    # labels : (No of Data Points,)
    def calculate_accuracy(self,dataset,labels):
        predictions = self.perform_classification(dataset)
        correct = 0
        for i in range(len(predictions)):
            if(predictions[i]==labels[i]):
                correct+=1
                    #print(correct)
        acc = ((correct*1.0)/len(predictions)* 100.0 ) 
        return acc
        
```

```python
def lda():
    accuracies = []
    for K in range(8,21):

        # Perform Principal Component Analysis
        pca = PCAHelper(train_data,K)    
        pca_train_data = pca.perform_pca()
        pca_test_data = pca.transform(test_data)
        
        # Perform LDA using LDA Helper class 
        lda = LDAHelper(pca_train_data,train.emotion)
        lda.perform_lda()
        
        # Calculate the accuracy of the LDA based classifier on the test set
        acc = lda.calculate_accuracy(pca_test_data,labels=test.emotion)
        accuracies.append(acc)
    
    # Return the accuracies obtained using LDA for different choices of K
    return accuracies

def svm(svm_class='sigmoid'):
    '''
       Produces a 3D plot showing the variation of accuracy for choices of tolerance and regularization
       parameter

       Parameters
        --------------
            svm_class: A string  variable representing the class for the kernel.
    '''
    # The value of the regularization parameter varies from 1 to 150 in increments of 3
    C = [i for i in range(1,150,5)]
    
    # The parameter choices for the tolerance value epsilon
    epsilon = [1e-2,1e-3,1e-4,2e-2,2e-3,2e-4,3e-1,3e-2,3e-3,4e-1,4e-2,4e-3]
    

    ret_accuracies = [] # stores tuples (K,maximum_accuracy)  for choices of K and maximum accuracy for each K

    for K in range(8,21):
        # Perform PCA to reduce the dimensionality of the data
        pca = PCAHelper(train_data,K)    
        pca_train_data = pca.perform_pca()
        pca_test_data = pca.transform(test_data)

        
        # Generate the axes for plotting the set of points
        fig = plt.figure()
        ax = plt.axes(projection ='3d')
        
        # maximum_accuracy : stores the maximum accuracy attained for a particular choice of K
        maximum_accuracy = 0
        
        # Iterate over the different values of tolerance 
        # and regularization parameter
        for i  in range(len(epsilon)):
            e = epsilon[i]
            
            # Dummy array for the purpose of plotting points on 3D
            epsilons = []
            # Array used for holding the values of the regularization parameter
            Cs = []

            # List holding accuracies corresponding to various choices of epsilon and C
            accuracies = []
            for c in C:
                # Construct the svm from the given regularization parameter and tolerance
                svm = SVC(C=c,  kernel=svm_class,tol=e)
                
                # Fit the SVM Model and use it for prediction
                svm.fit(pca_train_data,train.emotion)
                predictions = svm.predict(pca_test_data)

                # Calculating the accuracy by counting number of correctly classified points
                correct = 0
                for i in range(len(predictions)):
                    if(predictions[i]==test.emotion[i]):
                        correct+=1
                    #print(correct)
                acc = ((correct*1.0)/len(predictions)* 100.0 )   
                
                # Append to the different arrays which will be used for plotting
                Cs.append(c)
                epsilons.append(e)
                accuracies.append(acc)
            
            # Set the titles of the 3D plot
            ax.set_title("K = " + str(K)) 
            ax.set_xlabel("C")  
            ax.set_ylabel("Tolerance (epsilon)")
            ax.set_zlabel("Accuracy")
            
            # Keep track of the maximum accuracy out of a particular choice of K
            if(maximum_accuracy < max(accuracies)):
                maximum_accuracy = max(accuracies)

            ax.plot3D(Cs,epsilons,accuracies)
        print("Maximum accuracy attained for K = " + str(K) + " Kernel = " + svm_class + " = " + str(maximum_accuracy))
        # Append the maximum accuracy obtained for a particular choice of K and the kernel.
        ret_accuracies.append(maximum_accuracy)
        plt.show()
    
    return ret_accuracies
```

```python
linear_kernel_accuracies = svm('linear')
```

```python
sigmoid_kernel_accuracies = svm('sigmoid')
```

```python
polynomial_kernel_accuracies = svm('poly')
```

```python
lda_accuracies = lda()
```

#### Performance as a function of K for different Kernels

```python
K_values = [ i for i in range(8,21)]
plt.xlim([6, 30])
plt.xlabel("K")
plt.ylabel("Accuracy")
plt.plot(K_values,linear_kernel_accuracies,label="linear")
plt.plot(K_values,sigmoid_kernel_accuracies,label="sigmoid")
plt.plot(K_values,polynomial_kernel_accuracies,label="polynomial")
plt.plot(K_values,lda_accuracies,label="LDA")
plt.legend(loc="lower right")

plt.show()
```

### Observations

### 5a)

- The Linear Kernel gave 100% accuracy on the test set for K=11
- The sigmoidal Kernel has almost similar performance to the Linear Kernel as indicated in the graph above
- The polynomial kernel performed the poorest compared to other kernel based classifiers
- For a particular choice of the tolerance the performance of the linear kernel based classifier is constant
  throughout for different values of regularization coefficient. A higher value of tolerance improved accuracy in some cases in the linear kernel

- The plots of C vs epsilon in the polynomial kernel gave similar behaviour to linear kernel after choosing K >= 10
- Keeping the epsilon value constant and a particular choice of K, The sigmoidal kernel in some cases showed a shoot in accuracy followed by a decrease showing signs of underfitting with increasing values of C

- For some kernels giving a tolerance value {0.3,0.4} gave different plots for the (Accuracy vs Regularization parameter) graph compared to lower values of epsilon which gave similar plots


- We could see a constant accuracy with increasing values of K for the polynomial Kernel
- We can see fluctuating beahviours in accuracy with increasing values of K for the linear and sigmoid kernel
