```python
# Importing the required libraries
import os
import matplotlib.pyplot as plt
import PIL
import numpy as np
import math
```

```python
'''
Helper class for holding dataset
    *Attributes
        * data_type : holds the type of the dataset (train/test)
        * N : number of image files in the dataset
        * max_pixel_value : holds the maximum value pixel in the dataset
        * n_features : Dimension of the image vector (10201)
        * happy_indices: stores i such that files[i] is a happy images
        * sad_indices : similar to happy_indices for sad images    
        * dataset : a numpy matrix of dimensions = (Number of images) x (Dimensionality of each image) 
'''

class Dataset:
    
    def __init__(self,dataset_type="train/"):
        
        self.data_type = dataset_type
        self.files = os.listdir(dataset_type)
        self.N = len(self.files)
        self.max_pixel_value = []
        self.n_features = []
        self.happy_indices = []
        self.sad_indices = []
        self.dataset = None
    
    # Constructs the data matrix X of dimensions (Number of images x Dimension of 1 image vector)
    
    def getDataset(self):
        index = 0
        for file in self.files:
            pilimage = PIL.Image.open(self.data_type + file)
            img_class = file.split(".")[1]
            if(img_class=='sad'):
                self.sad_indices.append(index)
            else:
                self.happy_indices.append(index)
    
            pilimage.seek(0)
            tmp = pilimage.convert()
    
            current_image = np.asarray(tmp)
            current_image = current_image.reshape((1,-1))
            if self.dataset is None:
                self.dataset = current_image
            else:
                self.dataset = np.append(self.dataset,current_image,axis=0)
    
            self.max_pixel_value.append(np.max(current_image,axis=1)[0])
            self.n_features.append(current_image.shape[1])
            index+=1
        return self.dataset
```

```python
train = Dataset("train/")
N = train.N
dataset = train.getDataset()
```

```python
print("Size of the Training Dataset = " + str(N))
print("\nShape of Data Matrix = " + str(dataset.shape))
print("\nMaximum Pixel value in each image  = " + str(train.max_pixel_value)) # Checking whether images are gray scale
print("\nDimension of each image vector = " + str(train.n_features))
```

```python
def compute_mean_vector(dataset,N, feature_size):
    mean_vector = np.sum(dataset,axis = 0) * (1.0/N)
    return mean_vector

mean_vector = compute_mean_vector(dataset,train.N,10201).reshape((-1,10201))
print(mean_vector.shape)
```

```python
mean_centered_data = None
for i in range(len(dataset)):
    tmp = dataset[i]
    tmp = tmp - mean_vector
    if(mean_centered_data is  None):
        mean_centered_data = tmp
    else:
        mean_centered_data = np.append(mean_centered_data,tmp,axis=0)

print("X = " + str(mean_centered_data))
print("\nShape of the mean centered data = " + str(mean_centered_data.shape))
```

```python
xx_t = np.matmul(mean_centered_data,mean_centered_data.transpose())
xx_t = xx_t*(1.0/train.N)
xx_t
```

```python
eigen_value, eigen_vectors = np.linalg.eig(xx_t)
eigen_value
```

```python
indices = np.argsort(eigen_value)[::-1]
eigen_value = eigen_value[indices]
eigen_vectors = eigen_vectors[:,indices]
```

```python

component_1 = np.matmul(mean_centered_data.transpose(),v_1)
component_1 = component_1 * (1.0/math.sqrt(eigen_value[0]*train.N))
component_1
```

```python
x_t_x = np.matmul(mean_centered_data.transpose(),mean_centered_data)*(1.0/train.N)
x_t_x.shape
```

```python
e, v = np.linalg.eigh(x_t_x)
e
```

```python
indices = np.argsort(e)[::-1]
e = e[indices]
v = v[:,indices]
e
```

```python

v_2 = eigen_vectors[:,1]
v_2 = v_2.reshape((20,-1))

component_2 = np.matmul(mean_centered_data.transpose(),v_2)
component_2 = component_2 * (1.0/math.sqrt(eigen_value[1]*train.N))
component_2
```

```python
U = None

for i in range(eigen_vectors.shape[1]):
    eigen_vector = eigen_vectors[:,i].reshape((train.N,-1))# Each V_{i}
    #print(eigen_vector)
    ui = np.matmul(mean_centered_data.transpose(),eigen_vector)*(1.0/math.sqrt(np.abs(eigen_value[i])*train.N))
    if U is None:
        U = ui
    else:
        U = np.append(U,ui,axis=1) # U becomes a 10201*20 matrix

print(U) # Each u_{i} will be U[:,i] where 10201 is the dimension of each u_{i}
```

```python
K = 13
chosen_vectors = U[:,:K]
chosen_vectors.shape
```

```python
chosen_vectors
```

```python
projected_data = None
for i in range(N):
    data_point = mean_centered_data[i].reshape((-1,1))
    projection = np.matmul(chosen_vectors.transpose(),data_point)
    if projected_data is None:
        projected_data = projection
    else:
        projected_data = np.append(projected_data,projection,axis=1)

projected_data.shape
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=17, svd_solver='full')
pca.fit_transform(dataset)
```

```python
projected_data = projected_data.transpose()
projected_data.shape
```

```python

mean_happy = np.sum(projected_data[train.happy_indices],axis=0)*(1.0/len(train.happy_indices))
mean_sad = np.sum(projected_data[train.sad_indices],axis=0)*(1.0/len(train.sad_indices))
print(mean_happy)
print(mean_sad)
```

```python
happy_covariance_matrix = None
for i in range(len(train.happy_indices)):
    vector = (projected_data[train.happy_indices[i]] - mean_happy).reshape((-1,1))
    if happy_covariance_matrix is None:
        happy_covariance_matrix = (np.matmul(vector,vector.transpose()))*(1.0/len(train.happy_indices))
    else:
        happy_covariance_matrix += (np.matmul(vector,vector.transpose()))*(1.0/len(train.happy_indices))

print(happy_covariance_matrix.shape)
```

```python
sad_covariance_matrix = None
for i in range(len(train.sad_indices)):
    vector = (projected_data[train.sad_indices[i]] - mean_sad).reshape((-1,1))
    if sad_covariance_matrix is None:
        sad_covariance_matrix = (np.matmul(vector,vector.transpose()))*(1.0/len(train.sad_indices))
    else:
        sad_covariance_matrix += (np.matmul(vector,vector.transpose()))*(1.0/len(train.sad_indices))

print(sad_covariance_matrix.shape)
```

```python
Sw = happy_covariance_matrix + sad_covariance_matrix
Sw.shape
```

```python
Sw_inv = np.linalg.inv(Sw)
Sw_inv.shape
```

```python
projection_vector = np.matmul(Sw_inv,(mean_happy.reshape((-1,1))-mean_sad.reshape((-1,1))))
print(projection_vector)
```

```python
projection_u = []
for i in range(len(projected_data)):
    projection_value = np.matmul(projected_data[i],projection_vector)[0]
    projection_u.append(projection_value)
print(projection_u)
```

```python
max_value =  max(projection_u)
min_value = max(projection_u)
y = 0


count = 0 
for h in train.happy_indices:
    if(projection_u[h] < 0.0):
        count+=1;
    plt.plot(projection_u[h],y,'ro')

for s in train.sad_indices:
    if(projection_u[s]>=0.0):
        count+=1
    plt.plot(projection_u[s],y,'bo')

happy_mean_projection = np.matmul(mean_happy.reshape((1,-1)),projection_vector)[0]
sad_mean_projection = np.matmul(mean_sad.reshape((1,-1)),projection_vector)[0]
plt.plot(happy_mean_projection,y,'kp')
plt.plot(sad_mean_projection,y,'kp')

plt.plot([0, 0], [1, -1], marker = 'o')

plt.show()
count
```

```python
test_files = os.listdir("test")
test_dataset= None
sad_test = []
happy_test = [] 
index = 0


for file in test_files:
    #current_image = plt.imread("train/" + file)
    pilimage = PIL.Image.open("test/" + file)
    
    img_class = file.split(".")[1]
    if(img_class=='sad'):
        sad_test.append(index)
    else:
        happy_test.append(index)
    
    pilimage.seek(0)
    tmp = pilimage.convert()
    
    current_image = np.asarray(tmp)
    current_image = current_image.reshape((1,-1))
    if test_dataset is None:
        test_dataset = current_image
    else:
        test_dataset = np.append(test_dataset,current_image,axis=0)
    index+=1

print(test_dataset.shape)
```

```python
projected_test_data = np.matmul(test_dataset,chosen_vectors)
projected_test_data.shape
```

```python
projection_test_u = []
for i in range(len(projected_test_data)):
    projection_value = np.matmul(projected_test_data[i],projection_vector)[0]
    projection_test_u.append(projection_value)
print(projection_test_u)
```

```python
max_value =  max(projection_test_u)
min_value = max(projection_test_u)

misclassification_error = 0

for i in range(len(test_dataset)):
    if(projection_test_u[i]>= 0):
        predicted_class = 'happy'
    else:
        predicted_class = 'sad'
        
        
    if i in happy_test:
        actual_class ='happy' 
    else:
        actual_class = 'sad'
    
    if(predicted_class!=actual_class):
        misclassification_error+=1
        if(actual_class=='happy'):
            plt.plot(projection_test_u[i],y,'rx')
        else:
            plt.plot(projection_test_u[i],y,'bx')
    else:
        if (actual_class=='happy'):
            plt.plot(projection_test_u[i],y,'ro')
        else:
            plt.plot(projection_test_u[i],y,'bo')


plt.plot([0, 0], [1, -1], marker = 'o')
plt.show()   
misclassification_error
```
