```python
sample_rate, wav_file = wavfile.read("clean.wav")
print("Length of wav file = " + str(len(wav_file)))
length = len(wav_file)/sample_rate
print("Time Length of the wav file = " + str(length))
```

```python
class WhiteningTransform:
    
    def __init__(self):
        self.mean_vector = None
    
    def compute_spectrogram(self,wav_file):
        starting = 0
        increment = 160
        window_size = 400

        spectrogram = None
        while(starting+window_size <= len(wav_file)):
            current_window = np.array(wav_file[starting : starting + window_size])
            fft_transform = np.log(np.abs(np.fft.fft(current_window,n=256)[:128].reshape(128,1)))
            if spectrogram is None:
                spectrogram = fft_transform
            else:
                spectrogram = np.append(spectrogram,fft_transform,axis = 1)
    
            starting += increment

        print(spectrogram.shape)
        return spectrogram
    
    def plot_time_series_graph(self,wav_file):
        time = np.linspace(0,length,wav_file.shape[0])
        plt.plot(time, wav_file, label="channel")
        plt.show()

        
        
    def plot_spectrogram(self,spectrogram):
        time = np.linspace(0,311,num=311)
        freq = np.linspace(0,128,128)
        plt.pcolormesh(time,freq,spectrogram)
        plt.show()
        
        
    def compute_mean(self,spectrogram): 
        mean_vector = (np.sum(spectrogram,axis=1))*(1.0/spectrogram.shape[1])
        self.mean_vector = mean_vector
        return mean_vector
    
    def compute_mean_centered_data(self,spectrogram,mean_vector):
        mean_centered_data = spectrogram - mean_vector
        return mean_centered_data
    
    def compute_covariance_matrix(self,mean_centered_data):
        N = mean_centered_data.shape[1]
        covariance_matrix = (1.0/N)*(np.matmul(mean_centered_data,mean_centered_data.transpose()))
        return covariance_matrix
    
    def compute_eigen_values_eigen_vectors(self,covariance_matrix):
        eigen_values, eigen_vectors = np.linalg.eigh(covariance_matrix)
        # Sort the eigen values and vectors by decreasing eigen values

        indices = np.argsort(eigen_values)[::-1]
        eigen_values = eigen_values[indices]
        eigen_vectors = eigen_vectors[:,indices]
        print("\nLargest Eigen Value = " + str(eigen_values[0])) # Print the largest eigen value 
        print("\nEigen Vector corresponding to the largest eigen value\n\n" + str(eigen_vectors[:,1])) # The eigen vector corresponding to eigen value eigen_values[0]

        print("\nEigen Values in Descending Order\n")
        for i in range(len(eigen_values)):
            print(eigen_values[i])
        print("\n")
        return eigen_values,eigen_vectors
    
    def project_data(self,eigen_vectors,mean_centered_data):
        return np.matmul(eigen_vectors.transpose(),mean_centered_data)
    
    def compute_average_covariance_matrix(self,covariance_matrx):
        s = 0.0
        count = 0
        for i in range(covariance_matrix.shape[0]):
            for j in range(covariance_matrix.shape[1]):
                if (i!=j):
                    s += covariance_matrix[i][j]
                    count+=1
        
        return s/count
                    

        
w = WhiteningTransform()        
```

```python
w.plot_time_series_graph(wav_file)
```

```python
clean_spectrogram = w.compute_spectrogram(wav_file)
w.plot_spectrogram(clean_spectrogram)
```

```python
mean_vector = w.compute_mean(clean_spectrogram).reshape((-1,1))
mean_centered_data = w.compute_mean_centered_data(clean_spectrogram,mean_vector)
mean_vector
```

```python
covariance_matrix = w.compute_covariance_matrix(mean_centered_data)
covariance_matrix
```

```python
e,v = w.compute_eigen_values_eigen_vectors(covariance_matrix)
```

```python
projected_data = w.project_data(v,mean_centered_data)
projected_data
```

```python
diag_e = np.diag(e)
diag_e
```

```python
diag_inverse = np.linalg.inv(diag_e)
diag_inverse = np.sqrt(diag_inverse)
diag_inverse
```

```python
whitened_data = np.matmul(diag_inverse,projected_data)
whitened_data
```

```python
w.plot_spectrogram(whitened_data)
```
